```python
import pandas as pd

# Load the dataset
data = pd.read_csv("household_power_consumption_dataset.csv")
```

```python
data
```

```python

data['datetime'] = pd.to_datetime(data['Date'] + ' ' + data['Time'])

data.drop(['Date', 'Time'], axis=1, inplace=True)

# Display the first few rows of the processed data
print(data.head())
```

```python
data
```

```python
# Handle missing values
data = data.replace({'?': None})
data = data.fillna(method='ffill')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the color palette and edge color for the line plot
line_color = '#007F73'
edge_color = '#007F73'

# Trend of global active power over time
plt.figure(figsize=(15, 6))  # Increase figure width
sns.lineplot(x='datetime', y='Global_active_power', data=data, color=line_color, linewidth=2)
plt.title('Trend of Global Active Power Over Time', fontsize=18)
plt.xlabel('Date', fontsize=16)
plt.ylabel('Global Active Power', fontsize=16)
plt.xticks(fontsize=14)  # Adjust font size of x-axis ticks
plt.yticks(fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns  # Importing seaborn for advanced visualization options

# Assuming 'data' is a pandas DataFrame that contains the feature 'Global_active_power'
features_to_visualize = ['Global_active_power', 'Global_reactive_power', 'Voltage', 'Global_intensity', 'Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']

# Set a color palette for the histogram bars
color = '#4CCD99'

# Create individual histograms for each feature
for feature in features_to_visualize:
    plt.figure(figsize=(12, 7))  # Slightly larger figure for better readability
    sns.histplot(data[feature], bins=50, kde=True, color=color, edgecolor='#007F73')  # Using seaborn for kernel density estimate (KDE)
    plt.title(f"Distribution of {feature}", fontsize=18)  # Slightly larger title font size
    plt.xlabel(feature, fontsize=16)  # Larger x-axis label font size
    plt.ylabel('Frequency', fontsize=16)  # Larger y-axis label font size
    plt.grid(True, linestyle='--', alpha=0.6)  # Slightly darker grid lines for better visibility
    plt.xticks(fontsize=14)  # Larger x-axis tick font size
    plt.yticks(fontsize=14)  # Larger y-axis tick font size
    plt.tight_layout()  # Adjust layout to prevent overlap
    plt.show()  # Display the plot
```
